```python
from skimage.data import imread
```

```python
image = imread('data/basic.jpg')[::-1,::-1]
```

```python
imshow(image)
```

```python
from skimage.color import rgb2grey
```

```python
grey_image = rgb2grey(image)
```

```python
grey_image
```

```python
bin_image = grey_image < 0.5
```

```python
imshow(bin_image, cmap=cm.Greys)
```

```python
from skimage.segmentation import clear_border
```

```python
clear_image = clear_border(bin_image)
```

```python
imshow(clear_image, cmap=cm.Greys)
```

```python
from skimage.morphology import label
```

```python
label_image = label(clear_image, background=0)
```

```python
imshow(label_image)
```

```python
imshow(label_image==1)
```

```python
from skimage.measure import regionprops
```

```python
regions = regionprops(label_image, properties='all')
```

```python
char = regions[0]['Image']
```

```python
regions[0]
```

```python
imshow(char)
```

```python
from skimage.transform import resize
```

```python
resize_char = resize(char, (64,64), order=1)
```

```python
imshow(resize_char)
```

```python
bw_resize = rgb2grey(resize_char)
```

```python
imshow(bw_resize>0.5, cmap=cm.Greys)
```
